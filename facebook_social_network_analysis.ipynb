```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np
```

```python

G_fb = nx.read_edgelist(r"C:\Users\radhi\Desktop\project2\facebook_combined.txt\facebook_combined.txt", nodetype=int)
```

```python
print("Number of nodes:", G_fb.number_of_nodes())
print("Number of edges:", G_fb.number_of_edges())
```

```python
degrees = [d for n, d in G_fb.degree()]
plt.figure(figsize=(10, 6))
plt.hist(degrees, bins=range(1, max(degrees)+2), edgecolor='black')
plt.title('Degree Distribution (Facebook Network)')
plt.xlabel('Degree')
plt.ylabel('Number of Nodes')
plt.yscale('log')
plt.show()
```

```python
avg_clustering = nx.average_clustering(G_fb)
print("Average clustering coefficient:", avg_clustering)
```

```python
triangles = sum(nx.triangles(G_fb).values()) // 3
print("Number of triangles in the network:", triangles)
```

```python
largest_cc = max(nx.connected_components(G_fb), key=len)
```

```python

sample_cc_nodes = list(largest_cc)[:200] 
G_fb_sample = G_fb.subgraph(sample_cc_nodes)

diameter_sample = nx.diameter(G_fb_sample)
avg_shortest_path_sample = nx.average_shortest_path_length(G_fb_sample)

print("Sample Diameter (200 nodes):", diameter_sample)
print("Sample Avg Shortest Path (200 nodes):", avg_shortest_path_sample)
print("Sample size:", len(G_fb_sample))
```

```python

sample_cc_nodes = list(largest_cc)[:200]
G_fb_sample = G_fb.subgraph(sample_cc_nodes)
betweenness = nx.betweenness_centrality(G_fb_sample)
top_5 = sorted(betweenness.items(), key=lambda x: x[1], reverse=True)[:5]
print("Top 5 nodes by betweenness centrality (sample):", top_5)
```

```python
assortativity = nx.degree_assortativity_coefficient(G_fb)
print("Degree assortativity coefficient:", assortativity)
```
