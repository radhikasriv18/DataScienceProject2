```python
import networkx as nx
import matplotlib.pyplot as plt
import numpy as np


G_road = nx.read_edgelist(r"C:\Users\radhi\Desktop\project2\roadNet-TX.txt\roadNet-TX.txt", nodetype=int, create_using=nx.DiGraph(), comments='#')
```

```python
print("Number of nodes:", G_road.number_of_nodes())
print("Number of edges:", G_road.number_of_edges())
```

```python
in_degrees = [d for n, d in G_road.in_degree()]
out_degrees = [d for n, d in G_road.out_degree()]

plt.figure(figsize=(10, 6))
plt.hist(in_degrees, bins=range(1, max(in_degrees)+2), edgecolor='black')
plt.title('In-Degree Distribution (Texas Road Network)')
plt.xlabel('In-Degree')
plt.ylabel('Number of Nodes')
plt.yscale('log')
plt.show()

plt.figure(figsize=(10, 6))
plt.hist(out_degrees, bins=range(1, max(out_degrees)+2), edgecolor='black')
plt.title('Out-Degree Distribution (Texas Road Network)')
plt.xlabel('Out-Degree')
plt.ylabel('Number of Nodes')
plt.yscale('log')
plt.show()
```
